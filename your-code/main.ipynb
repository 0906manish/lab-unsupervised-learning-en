```python
import numpy as np

def relevant_column(arr1, arr2):
    
    arr1 = np.array(arr1)
    arr2 = np.array(arr2)
    
    if len(arr1) != len(arr2):
        return "Size mismatch"
    
    total_sum = np.sum((arr1 - arr2)**2)

    if total_sum < 0.1:
        return "Highly correlated / Redundant (Drop one)"
    elif total_sum > 1.5:
        return "Likely independent / Distinct features"
    else:
        return "Partial dependency"


result = relevant_column(df.iloc[:, 0].tolist(), df.iloc[:, 1].tolist())
print(result)
```
